```python
import pandas as pd
from bs4 import BeautifulSoup
import requests

# Scraping French top 14 Standings from bbc.com, avoid manual labour at all costs!

curr_url = "https://www.bbc.com/sport/rugby-union/top-14/table"
page = requests.get(curr_url)
soup = BeautifulSoup(page.text, 'html.parser')

dfs = pd.read_html(page.text)

print(dfs)

for child in soup.find_all('table')[0].children:
    print(child)
    print()

```
